```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import numpy as np
from sklearn.cluster import KMeans
from sklearn import preprocessing
import pandas as pd
```

```python
df = pd.read_csv('Harukei-3_30tissues_gene_expression_corrected_FPKM_2017_ver1.csv',delimiter=',')
```

```python
df.head()
```

```python
gene_ids = df['gene_id']
df.drop(['gene_id'], 1, inplace=True)
```

```python
df.head()
```

```python
X = np.array(df.astype(float))
X = preprocessing.scale(X)
```

```python
from scipy.spatial.distance import cdist

# k means determine k using elbow method
distortions = []
K = range(2,200,1)
for k in K:
    #kmeanModel = KMeans(n_clusters=k).fit(X)
    #kmeanModel.fit(X)
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    distortion_value = sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0]
    print("K = ",k,", Distortion value = ",distortion_value)
    distortions.append(distortion_value)

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score

range_n_clusters = range(10,20001,500)

for my_n_clusters in range_n_clusters:
    clf = KMeans(n_clusters=my_n_clusters)
    cluster_labels = clf.fit_predict(X)
    
    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", my_n_clusters,
          "The average silhouette_score is :", silhouette_avg)

```

```python
clf = KMeans(n_clusters=1000)
clf.fit(X)

for i in range(len(X)):
    predict_me = np.array(X[i].astype(float))
    predict_me = predict_me.reshape(-1, len(predict_me))
    prediction = clf.predict(predict_me)
    print("{}\t{}".format(gene_ids[i],prediction[0]),file=open("Harukei-3_30tissues_gene_expression_corrected_FPKM_2017_ver1_KMeans_clusters.txt", "a"))
```
